```python
df = pd.read_csv('Salaries.csv')
df.head()
```

### What is an average base pay?

```python
df.replace(["Not Provided", 'Not provided'], np.nan, inplace = True)
df["BasePay"].astype(float).mean()
```

### How many Job Titles were represented by only one person in 2013 (eg. Job Titles was only one occurance in 2013)?

```python
one_person = df['JobTitle'].value_counts()==1
type(one_person)
len(one_person)
counter =0
for i in range(0, len(one_person)):
    if one_person[i]== True:
        counter+=1
    
print(counter)
```

### How many people have the word Chief in their job title? 

```python
df1 = df.copy()

df2 = df1['JobTitle']
df2
df2.count
```

```python
df2.dropna(inplace = True)
counter = 0;
for i in range(0, len(df2)):
    if 'CHIEF' in df2.iloc[i].split() or 'chief' in df2.iloc[i].split():
        counter +=1
        continue
print(counter)
```
